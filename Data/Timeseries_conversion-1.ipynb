```python
# adding all the sequences
#using import csv lib
start=0

f = open('data1.csv')
csv_f = csv.reader(f)
with open("data1_Numeric_csv.txt",'w') as WF:
    for row in csv_f:
        print (row[1])
        WF.write("\n")
        for x in row[1]:
            print ('x=',x) #printing one letter one by one
            if x in vals:
                    print (vals[x]) # converting only the capital letters of the word to a number
                    print (vals[x],"+",start,"=",start+vals[x]) #printing the added values
                    start+=vals[x] #adding the converted numbers
            else:
                    print (start)
                    print ("entered else:",start ,"+", vals["_"],"=",start+vals["_"]) # handling encounter of other letters 
                    start+=vals["_"]
            WF.write(str(start)+" ")
```
